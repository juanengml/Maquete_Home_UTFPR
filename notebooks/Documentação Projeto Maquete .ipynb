## Problematica: 
Condições Pertinentes

* Simulação de ambientes de luminotecnica
* Simulação de problemas em pesquinas escalas
* Testes automatizados em rotinas pré-definida
* Controle de Ambientes usando plataformas web
* Teste de rotinas de automação

## Objetivo
* Controle de Ambiente usando técnologia de baixo custo
* Integração com Nuvem 

## Prototipos 
* Contrução e Elaboração da Maquete em conjunto com Engenharia Civil
* Usando arduino e shield ethernet para controle e simulação  
* Integração e testes da eletronica usando thingsboard como plataforma detest
